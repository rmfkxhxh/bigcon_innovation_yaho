```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
```

```python
from sklearn.cluster import KMeans
model_kpp = KMeans(n_clusters=3, init="k-means++")
```

```python
retail_data = pd.read_pickle('../../data/local_data/pkls/cat_cost.pickle')
retail_data.columns
```

```python
total_cost = retail_data.groupby(['ADMD_NM', 'date', 'ANTC_ITEM_LCLS_NM'])[['Total Cost']].mean().unstack(-1).groupby('ADMD_NM').mean()
```

```python
cat_cost = retail_data.groupby(['ADMD_NM', 'date', 'ANTC_ITEM_LCLS_NM'])[['cat_cost']].mean().unstack(-1).groupby('ADMD_NM').mean()
```

```python
x = pd.concat([cat_cost, total_cost],axis=1)
```

```python
scaler=StandardScaler()
scaler.fit(x)
x1=scaler.transform(x)
model_kpp.fit(x1)
model_kpp.cluster_centers_
x=pd.concat([x.reset_index(),pd.Series(model_kpp.labels_)],axis=1)
```
