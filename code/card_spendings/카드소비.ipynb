```python
def plot_multi(data, cols=None, spacing=.1, **kwargs):

    from pandas import plotting

    # Get default color style from pandas - can be changed to any other color list
    if cols is None: cols = data.columns
    if len(cols) == 0: return
    colors = getattr(getattr(plotting, '_style'), '_get_standard_colors')(num_colors=len(cols))

    # First axis
    ax = data.loc[:, cols[0]].plot(label=cols[0], color=colors[0], **kwargs)
    ax.set_ylabel(ylabel=cols[0])
    lines, labels = ax.get_legend_handles_labels()

    for n in range(1, len(cols)):
        # Multiple y-axes
        ax_new = ax.twinx()
        ax_new.spines['right'].set_position(('axes', 1 + spacing * (n - 1)))
        data.loc[:, cols[n]].plot(ax=ax_new, label=cols[n], color=colors[n % len(colors)])
        ax_new.set_ylabel(ylabel=cols[n])

        # Proper legend position
        line, label = ax_new.get_legend_handles_labels()
        lines += line
        labels += label

    ax.legend(lines, labels, loc=0)
    return ax
```

```python
# 1. 자료 수집 코드

import tweepy
from pandas import DataFrame as df
from sys import argv

api_key = "LCB6MOev6MePyKM5bAzyxS0gd" 
api_secret = "OPMhFwYJreOh8xWZbxV7nhMKZYSfiCCcPBOw8HkmDI8eMaIm0u" 
auth = tweepy.OAuthHandler(api_key, api_secret)

api = tweepy.API(auth) 

dates = []
authors = []
ids = []
contents = []

keyword = r'노원구'

for tweetObj in tweepy.Cursor(api.search, q=keyword, count=100).items():
    tweetStr = tweetObj.text 
    tweetStr = tweetStr.replace("\n"," ")
    createdAt = tweetObj.created_at
    authorObj = tweetObj.author
    authorName = authorObj.screen_name
    tweetId = tweetObj.id
    print (tweetId, createdAt, authorName, tweetStr)
    ids.append(tweetId)
    dates.append(createdAt)
    authors.append(authorName)
    contents.append(tweetStr)

data = {'ID':ids, 'Date':dates, 'Author':authors, 'Content':contents}
tweets = df(data)
tweets.to_csv ('노원구.csv', header = True, encoding = 'utf-8')

# 2. 자료 전처리 및 분석 코드

#데이터 불러오기
import pandas as pd
misogyny = pd.read_csv("노원구.csv")
misogyny = misogyny[['ID', 'Date', 'Author', 'Content']]
misogyny = misogyny.drop_duplicates(['Author', 'Content'])

#데이터 처리 속도 향상을 위하여 트윗 내용을 여러 그룹으로 분할하여 딕셔너리에 저장
splitted = {}
for i in range(1,127):
    group = list(misogyny['Content'][100*(i-1):100*i])
    splitted[i] = ','.join(group)
splitted[127] = ','.join(list(misogyny['Content'][12600:]))

#트윗 내용에서 리트윗 아이디, 링크 주소, 이모티콘과 같은 정보 제거
import re
stripped = {}
for j in range(1,128):
    i = re.sub(r'@[a-zA-Z0-9_]*', ' ', str(splitted[j]))
    i = re.sub(r'https://[a-zA-Z0-9./?]*', ' ', i)
    i = re.sub(r'RT', ' ', i)
    i = re.sub(r'(ㅋ)+', ' ', i)
    Re_emoji = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)
    i = Re_emoji.sub(r'', i)
    stripped[j] = i

#Okt 라이브러리를 통하여 형태소 분석하여 명사 추출
from konlpy.tag import Okt
okt = Okt()
tags = {}
for i in range(1, 128):
    tags[i] = okt.nouns(stripped[i])

from functools import reduce
taglist = reduce (lambda x, y: x+y, tags.values())

#stopwords 제거
stopwords = ['것', '이', '저', '나', '걸', '거', '세', '그', '함', '내', '수', '게', '제']
remove = []
for i in taglist:
    if i not in stopwords:
        remove.append(i)

#단어 빈도 통계
from collections import Counter
import random
from konlpy.tag import Hannanum
import pytagcloud

r = lambda: random.randint(0,255)
color = lambda: (r(), r(), r())

import math
def get_tags(nouns, ntags=50, multiplier=30):
    count = Counter(nouns)
    return [{ 'color': color(), 'tag': n, 'size': c//multiplier }\
                for n, c in count.most_common(ntags)]

tagged = get_tags(remove)

#빈도 결과 정리
word = []
count = []
for i in tagged:
    word.append(i['tag'])
    count.append(int(i['size'])*30)

data = {'Word':word, 'Count':count}
wordcount = pd.DataFrame(data)
wordcount.to_csv ('wordcount.csv', header = True, encoding = 'utf-8')



#워드클라우드 만들기
import webbrowser

def draw_cloud(tags, filename, fontname='NanumBarunGothic', size=(800, 800)):
    pytagcloud.create_tag_image(tags, filename, fontname=fontname, size=size)
    webbrowser.open(filename)

draw_cloud(tagged, 'misogyny.png')
```
