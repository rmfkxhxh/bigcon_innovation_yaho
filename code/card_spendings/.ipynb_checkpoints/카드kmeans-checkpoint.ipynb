```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
```

```python
from sklearn.cluster import KMeans
model_kpp = KMeans(n_clusters=3, init="k-means++")
```

```python
card_data=pd.read_pickle('finaldata.pkl')
card_total=pd.read_pickle('total.pkl')
```

```python
card_total['TEMP']=card_total['USE_AMT']/card_total['USE_CNT']
```

```python
x=card_total.groupby(['STD_DD','DONG_CD','MCT_CAT_CD'])['USE_AMT'].sum().unstack(-1).groupby('DONG_CD').mean()
```

```python
x
```

```python
x=x.fillna(0)
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
```

```python
x1=scaler.transform(x)
```

```python
model_kpp.fit(x1)
```

```python
model_kpp.cluster_centers_
```

```python
x=pd.concat([x.reset_index(),pd.Series(model_kpp.labels_)],axis=1)
```

```python
x.columns
```

```python
x.columns=['DONG_CD', '레저용품', '보건위생', '서적문구', '음료식품', 'cluster']

```

```python
x_np=np.array(x)
```

```python
labels = model_kpp.predict(x_np)
centroids = model_kpp.cluster_centers_
```
