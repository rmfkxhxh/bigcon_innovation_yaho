```python
data=pd.read_pickle('news_10total_class')
```

```python
#일별 기준으로 카운트하기 위한 DATE칼럼 전처리
data.DATE=pd.to_datetime(data.DATE,format='%Y%m%d%H%M%S').dt.date
data.DATE=pd.to_datetime(data.DATE)
```

```python
#TOPIC(1,3,5,9,10)을 DATE기준으로 그룹화 해서 카운트(일별 카운트)
data[data.topic.isin([1,3,5,9,10])].groupby(['DATE']).count()
```

```python
#날짜 X축 topic갯수 Y축 으로 그린다
data[data.topic.isin([1,3,5,9,10])].groupby(['DATE']).count().reset_index().plot(kind='line',x='DATE',y='topic')
```

```python
#공휴일 기준으로 그린거
data[(data.topic.isin([1,3,5,9,10]))&(data.DATE.isin(holiday_df))].groupby(['DATE']).count().reset_index().plot(kind='line',x='DATE',y='topic')
```

```python
#평일만
data[(data.topic.isin([1,3,5,9,10]))&(~data.DATE.isin(holiday_df))].groupby(['DATE']).count().reset_index().plot(kind='line',x='DATE',y='topic')
```
