```python
import pandas as pd
import numpy as np
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt 
import seaborn as sns  
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
dong='사직동'
movedata=pd.read_pickle('movedata_worker')
carddata=pd.read_pickle('card_hh_1year')
movedata=movedata[movedata['HDONG_NM']==dong][['STD_YMD','S-W']].reset_index()
del movedata['index']
movedata.STD_YMD=pd.to_datetime(movedata.STD_YMD,format='%Y%m%d')
result=carddata.merge(movedata,how='inner',left_on='date',right_on='STD_YMD')
del result['date']
del result['pm25']
del result['pm10']
del result['pm_class']
del result['STD_YMD']

plt.figure(figsize=(15,15))
sns.heatmap(data = result.corr(method = 'pearson'), annot=True, 
fmt = '.2f', linewidths=.5, cmap='RdBu')
```
