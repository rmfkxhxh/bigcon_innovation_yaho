```python
import pandas as pd
```

```python
data_dust=pd.read_pickle('dust_hour')
data_dust=data_dust.reset_index()
data_move=pd.read_pickle('movedata_worker')
```

```python
data_move.STD_YMD=pd.to_datetime(data_move.STD_YMD,format='%Y%m%d')
data_dust.tm=pd.to_datetime(data_dust.tm,format='%Y-%m-%d')
```

```python
merge_df=pd.merge(data_dust,data_move,how='inner',left_on=['tm','serial'],right_on=['STD_YMD','HDONG_NM'])
```

```python
merge_df
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns    
%matplotlib inline  
plt.figure(figsize=(30,30))
sns.heatmap(data = merge_df.corr(method = 'pearson'), annot=True, 
fmt = '.2f', linewidths=.5, cmap='RdBu')
```
