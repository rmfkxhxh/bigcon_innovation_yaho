```
#불용어 불러오기 나중에 db에서 불러오기


#불용어 파일 불러오기
ksw = open('/content/drive/My Drive/Colab Notebooks/sns뿌시자/kstopword.txt', 'r',encoding='cp949')
read_ksw = ksw.read()

#리스트로 만둘기
split_ksw = read_ksw.split()

#중복제거
now_total=new_total.drop_duplicates(['CONTENT'])

#nan값 제거
now_total=new_total.dropna()

#제너레이터 만드는 함수
def content_gen(gene):
    return (i for i in tqdm_notebook(gene))
def gen(gene):
    return (gene)

#np로 담기
# temp_text=np.array(new_total[new_total['SECTION']=='카페']['CONTENT'])

#mecab성능이 좋아서 한번에 돌려보기
temp_text=np.array(new_total['CONTENT'])


#제너레이터 만들기
sns_text = content_gen(temp_text)
#한글 빼도 다제거 싹다제거 !!제너레이터는 for 문에 범위값에 넣어서 활용한다 i로 하나씩 불려와서 사용됨 
#한번쓰면 없어짐 다시선언해야됨
#어쩃든 fix_test는 한글로만된text list
fix_text=[[re.sub('[^가-힣]',' ',str(i))] for i in tqdm_notebook(sns_text)]

temp=pd.DataFrame(fix_text)
temp=np.array(temp)
del temp_text
```
